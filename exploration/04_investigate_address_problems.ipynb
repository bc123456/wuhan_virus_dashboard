```python
import pickle
import pandas as pd
```

```python
address_df_path = r'../data/ADDRESS.pkl'
address_old_df_path = r'../data/ADDRESS_old.pkl'
high_risk_df_path = r'../data/HIGH_RISK.pkl'
high_risk_old_df_path = r'../data/HIGH_RISK_old.pkl'
```

```python
with open(address_df_path, 'rb') as f:
    address_df = pickle.load(f)
with open(address_old_df_path, 'rb') as f:
    address_old_df = pickle.load(f)
    
with open(high_risk_df_path, 'rb') as f:
    high_risk_df = pickle.load(f)
    
with open(high_risk_old_df_path, 'rb') as f:
    high_risk_old_df = pickle.load(f)
```

```python
address_df.shape
```

```python
address_old_df.shape
```

```python
address_df = address_df.drop_duplicates(subset=['location_en', 'longitude', 'latitude'])
```

```python
high_risk_df.head()
```

```python
high_risk_old_df.head()
```

```python
pd.merge(
    high_risk_df,
    high_risk_old_df,
    how='left',
    left_on='id',
    right_on='id'
)[['location_zh_x', 'location_zh_y']]
```

```python
high_risk_with_coordinates_df = pd.merge(
	high_risk_df,
	address_df,
	how='left',
	left_on='location_en',
	right_on='location_en'
)
```
